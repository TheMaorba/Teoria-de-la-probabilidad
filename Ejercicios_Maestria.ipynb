+ **Problema**. _Un tirador debe acertar dentro de un circulo de radio 3 contenido en una region cuadrada de lado 20. ¬øCual es la Probailidad de que el Tirador acierte?_

```python

import matplotlib.pyplot as plt
import numpy as np

# Par√°metros
s = 8
r = 3

# Visualizaci√≥n
plt.figure(figsize=(6, 6))
plt.plot([0, s], [0, s], 'w-')  # L√≠nea blanca para ocultar los bordes del cuadrado
plt.xlim(0, s)
plt.ylim(0, s)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Probabilidad Geom√©trica: Punto en un √Årea")

# Dibujar el cuadrado
plt.gca().add_patch(plt.Rectangle((0, 0), s, s, edgecolor='black', fill=False))

# Dibujar el c√≠rculo
circle = plt.Circle((s / 2, s / 2), r, color='red', fill=False)
plt.gca().add_patch(circle)

# Agregar texto
plt.text(s / 2, s * 0.9, f"√Årea Cuadrado = {s**2}", ha='center')
plt.text(s / 2, s * 0.1, f"√Årea C√≠rculo = {np.round(np.pi * r**2, 2)}", ha='center', color='red')

# Probabilidad
probability = np.round(np.pi * r**2 / s**2, 3)
print(f"P(A) = {probability}")

plt.show()
```




## üìò **2. Problema de Probabilidad: Encuentro de dos amigos**

**Enunciado:**
_Dos amigos acuerdan encontrarse en un lugar entre las 12:00 PM y la 1:00 PM. Cada uno llega en un momento aleatorio dentro de esa hora (es decir, sus tiempos de llegada est√°n distribuidos uniformemente entre 0 y 60 minutos). El primero que llega esperar√° **15 minutos** al otro, despu√©s de lo cual se ir√°. ¬øCu√°l es la probabilidad de que se encuentren?_

---

## ‚úçÔ∏è **Soluci√≥n**
Sea:

* $X$ el tiempo de llegada del primer amigo.
* $Y$ el tiempo de llegada del segundo amigo.

Ambos tiempos est√°n distribuidos uniformemente en el intervalo $[0, 60]$ minutos.
La pareja se encuentra **si la diferencia entre sus tiempos de llegada es menor o igual a 15 minutos**, es decir:


$$
|X - Y| \leq 15
$$

### üî∑ **Representaci√≥n gr√°fi

Podemos representar este problema en el plano XY, donde $X$ y $Y$ var√≠an de 0 a 60. La regi√≥n total del espacio de posibles llegadas es un cuadrado de √°rea:

$$
A_{\text{total}} = 60 \times 60 = 3600
$$

La regi√≥n donde los amigos se encuentran est√° limitada por:

$$
|X - Y| \leq 15 \Rightarrow -15 \leq X - Y \leq 15
$$

Esto corresponde a la franja entre las rectas $Y = X - 15$ y $Y = X + 15$, dentro del cuadrado.

### üî∏ **√Årea favorable**

El √°rea de la regi√≥n donde **se encuentran** es:

$$
A_{\text{favorable}} = A_{\text{total}} - 2 \cdot \text{√Årea de los tri√°ngulos fuera de la banda}
$$

Cada tri√°ngulo tiene base y altura de 45 (desde 0 a 15 y de 45 a 60):

$$
\text{√Årea de un tri√°ngulo} = \frac{1}{2} \cdot 45 \cdot 45 = 1012.5
$$

Entonces:

$$
A_{\text{favorable}} = 3600 - 2 \cdot 1012.5 = 3600 - 2025 = 1575
$$

### üîπ **Probabilidad de encuentro**

Finalmente, la probabilidad es:

$$
P(\text{encuentro}) = \frac{A_{\text{favorable}}}{A_{\text{total}}} = \frac{1575}{3600} = \frac{7}{16} \approx 0.4375
$$

---

## ‚úÖ Respuesta final

**La probabilidad de que los dos amigos se encuentren es:**

$$
\boxed{\frac{7}{16} \approx 43.75\%}
$$

```python
# @title **Resolviendo en Python**
import matplotlib.pyplot as plt
import numpy as np

# Crear una grilla de puntos (X, Y) entre 0 y 60
x_vals = np.linspace(0, 60, 300)
y_vals = np.linspace(0, 60, 300)
x, y = np.meshgrid(x_vals, y_vals)

# Clasificar los puntos: dentro o fuera de la zona de encuentro
encuentro = np.where(np.abs(x - y) <= 15, "Se encuentran", "No se encuentran")

# Graficar
plt.figure(figsize=(8, 6))
plt.imshow(np.where(encuentro == "Se encuentran", 1, 0),
           extent=[0, 60, 0, 60], origin='lower', cmap='Blues', alpha=0.7)  # Usa un mapa de color azul

plt.plot([0, 60], [15, 75], linestyle='--', color='blue')  # Linea discontinua azul
plt.plot([0, 60], [-15, 45], linestyle='--', color='blue')

plt.title("Regi√≥n de encuentro de dos amigos")
plt.xlabel("Minuto en que llega el primer amigo")
plt.ylabel("Minuto en que llega el segundo amigo")

plt.colorbar(label="Encuentro")  # Agrega la barra de color

plt.grid(True)
plt.show()
```

```python
# @title **Resolviendo en Python**
import matplotlib.pyplot as plt
import numpy as np

# Crear una grilla de puntos (X, Y) entre 0 y 60
x_vals = np.linspace(0, 60, 300)
y_vals = np.linspace(0, 60, 300)
x, y = np.meshgrid(x_vals, y_vals)

# Clasificar los puntos: dentro o fuera de la zona de encuentro
encuentro = np.where(np.abs(x - y) <= 15, "Se encuentran", "No se encuentran")

# Graficar
plt.figure(figsize=(8, 6))
plt.imshow(np.where(encuentro == "Se encuentran", 1, 0),
           extent=[0, 60, 0, 60], origin='lower', cmap='Blues', alpha=0.7)  # Usa un mapa de color azul

plt.plot([0, 60], [15, 75], linestyle='--', color='blue')  # Linea discontinua azul
plt.plot([0, 60], [-15, 45], linestyle='--', color='blue')

plt.title("Regi√≥n de encuentro de dos amigos")
plt.xlabel("Minuto en que llega el primer amigo")
plt.ylabel("Minuto en que llega el segundo amigo")

plt.colorbar(label="Encuentro")  # Agrega la barra de color

plt.grid(True)
plt.show()


# Calculate the area of the favorable region and total region
total_area = 60 * 60  # Total area of the square
triangle_area = 0.5 * 45 * 45  # Area of one of the two triangles
favorable_area = total_area - 2 * triangle_area

# Calculate the probability
probability = favorable_area / total_area

# Print the results
print(f"Total area: {total_area}")
print(f"Area de un triangulo: {triangle_area}")
print(f"Favorable area: {favorable_area}")
print(f"Probabilidad pedida: {probability} or {probability*100:.2f}%")
```
