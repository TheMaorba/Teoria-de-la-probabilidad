# **a. Ejemplo: Simular el lanzamiento de dos dados $n = 100$ veces - Usando Python**

```python
# **a. Ejemplo: Simular el lanzamiento de dos dados n veces - Usando Python**
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
n = int(input("Escriba el número de veces en que lanzara los dos dados: "))
a = 1 # Valor minimo en un dado
b = 6 # Valor máximo en un dado
x=1 # nos va indicar el primer lanzamiento
dado1 = [] # Queremos colocar el resultado obtenido en los lanzamientos del primer dado
dado2 = [] # Queremos colocar el resultado obtenido en los lanzamientos del segundo dado
suma12 = [] # Queremos colocar el total de los resultados obtenido en los lanzamientos de los dos dados
media = [] # Queremos colocar el resultado promedio obtenido en los lanzamientos de ambos dados
varianza = [] # Queremos colocar la varianza de los resultados promedio obtenido en los lanzamientos de ambos dados
suma = 0 # Acumular los resultados sumandolos

while x<=n: # Esto es para decir que queremos lanzar el dado 10 veces
    valor1=random.randint(a,b) # Cada vez que se henere un lanzamiento, se genera un numero aleatorio entre 1 y 6
    dado1.append(valor1) # Me dice el resultado del dato
    valor2=random.randint(a,b) # Cada vez que se henere un lanzamiento, se genera un numero aleatorio entre 1 y 6
    dado2.append(valor2) # Me dice el resultado del dato
    suma12.append(valor1+valor2)
    media.append((valor1+valor2)/2)
    varianza.append((valor1-valor2)**2/2)

    suma=suma+valor1+valor2 # Acumulando la suma de los valores obtenidos


    x=x+1 # el ciclo se cierra y comienza
promedio=suma/n #Deseo hallar la media de los resultados de los 10 lanzamientos

dados = pd.DataFrame({'dado1':dado1,'dado2':dado2,'Total':suma12,'media':media,'varianza':varianza})


print(f'Los resultados de los {n} lanzamientos del primer dado fueron',
      dado1)
print(f'Los resultados de los {n} lanzamientos del segundo dado fueron',
      dado2)
print(f'Los resultados de los totales de {n} lanzamientos de los dados fueron',
      suma12)
print("La suma de los 10 valores es igual a ", suma)
print("El promedio de los resultados obtenidos fueron", promedio)
print(promedio)
print('En una tabla el resumen de nuestro experimento Tabla1. Dados')
print(dados)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Supongamos que 'dados' es tu DataFrame y ya tienes cargados los datos pertinentes

# Definir las columnas de interés
cols = dados.columns[2:]

# Número de columnas
N = len(cols)

# Crear figura y subplots
plt.figure(figsize=(25, 6))

# Para cada columna, ajustar una distribución normal y trazar la curva PDF
for i, col in enumerate(cols):
    plt.subplot(1, N, i+1)
    sns.histplot(dados[col], kde=True, stat="density", kde_kws=dict(cut=3),
                 alpha=.4, edgecolor=(1, 1, 1, .4))

    # Ajustar una distribución normal a los datos
    mu, std = norm.fit(dados[col])

    # Calcular la PDF de la distribución normal ajustada
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)

    # Trazar la curva de densidad de probabilidad (PDF)
    plt.plot(x, p, 'k', linewidth=2)

    # Configuración del gráfico
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Densidad de probabilidad')

plt.tight_layout()
plt.show()
```

### **a. FMP - Función de Masa de Probabilidad**
+ **Ejemplo.** Construyeme la funcion de masa de probabilidad para el lanzamiento de dos monedas y hazme el grafico

```python
# prompt: Construyeme la funcion de masa de probabilidad para el lanzamiento de dos monedas y hazme el grafico

import matplotlib.pyplot as plt
import numpy as np

# Definir el espacio muestral
muestra = np.array([0, 1, 2])

# Definir la función de masa de probabilidad
probabilidad = np.array([1/4, 1/2, 1/4])

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.bar(muestra, probabilidad, color='blue')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.title('Función de Masa de Probabilidad para el Lanzamiento de Dos Monedas')
plt.show()
```

## **Ejemplo. Construyeme la funcion de densidad de probabilidad para una variable exponencial de parametro 2 y hazme el grafico**

```python
# prompt: Construyeme la funcion de densidad de probabilidad para una variable exponencial de parametro 2 y hazme el grafico

import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the exponential distribution parameters
lambda_value = 2

# Generate the random variable
x = np.linspace(0, 10, 100)

# Calculate the probability density function (PDF)
y = stats.expon.pdf(x, scale=1/lambda_value)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue')
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Exponential Distribution PDF (lambda = {})'.format(lambda_value))
plt.show()
```

## **Ejemplo. Simulación del lanzamiento de dos dados donde se muestre la distribución o histograma de la media y la varianza**

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de la simulación
np.random.seed(42)  # Semilla para reproducibilidad
num_lanzamientos = 1000000

# Lanzamiento de dos dados
dados = np.random.randint(1, 7, size=(num_lanzamientos, 2))

# Calculando la media y la varianza para cada par de lanzamientos
medias = np.mean(dados, axis=1)
varianzas = np.var(dados, axis=1)

# Graficando histogramas de la media y la varianza
plt.figure(figsize=(12, 6))

# Histograma de la media
plt.subplot(121)
plt.hist(medias, bins=np.arange(1.5, 6.6, 1), edgecolor='black', alpha=0.7)
plt.title('Histograma de la Media de Dos Dados')
plt.xlabel('Valor Medio')
plt.ylabel('Frecuencia')

# Histograma de la varianza
plt.subplot(122)
plt.hist(varianzas, bins='auto', edgecolor='black', alpha=0.7)
plt.title('Histograma de la Varianza de Dos Dados')
plt.xlabel('Varianza')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

```
