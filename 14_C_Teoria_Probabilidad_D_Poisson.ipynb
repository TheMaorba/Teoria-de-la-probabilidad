## **a. La Definición**

+ La distribución de Poisson es una distribución de probabilidad discreta que expresa, a partir de una frecuencia de ocurrencia media, la probabilidad de que ocurra un determinado número de eventos durante cierto período de tiempo.
+ También puede utilizarse para contar el número de eventos o partículas consideradas puntuales en otros tipos de intervalos específicos, como la distancia, el área o el volumen.

+ Fue propuesta por *Siméon-Denis Poisson*, que la dio a conocer en 1838 en su trabajo *Investigación sobre la probabilidad de los juicios en materias criminales y civiles*.

+ Una variable aleatoria discreta $X$ sigue una distribución de Poisson si tiene funcion FMP dada por la formula:
$$ P(X=x)= f(x,λ) = e^{-λ}\frac{λ^x}{x!},  x≥0$$




## **b. Valor Esperado y Varianza de una variable aleatoria discreta de Bernoulli**

+ **Valor Esperado** $$E(X)=\mu=\sum_{x=o}^{∞}x e^{-λ}\frac{λ^x}{x!}$$
+ Se tiene que
$$E(X)=e^{-λ}\sum_{x=o}^{∞}x \frac{λ^x}{x!} \ \
=e^{-λ}\sum_{x=o}^{∞}x \frac{λ^x}{x(x-1)!}
 \ \
=λe^{-λ}\sum_{x=1}^{∞} \frac{λ^{x-1}}{(x-1)!}$$

+ Haciendo el cambio $k=x-1$

$$E(X)=λe^{-λ}\sum_{k=0}^{∞} \frac{λ^k}{k!}=λe^{-λ}e^{λ}=λ$$

+ **Reorganizando**

$$E(X)=\mu=λ$$



## **c. Varianza** $$V(X)=E(X-μ)^2=σ^2=E(X^2)- {\mu}^2$$


$$E(X^2)=\sum_{x=o}^{n}x^2e^{-λ}\frac{λ^x}{x!} \ \ =\sum_{x=o}^{n}[x(x-1)+x]e^{-λ}\frac{λ^x}{x!} $$

$$E(X^2)=\sum_{x=o}^{n}[x(x-1)]e^{-λ}\frac{λ^x}{x!} +\sum_{x=o}^{n}[x]e^{-λ}\frac{λ^x}{x!}$$

$$E(X^2)=\sum_{x=o}^{n}[x(x-1)]e^{-λ}\frac{λ^x}{x(x-1)(x-2)!} +\sum_{x=o}^{n}xe^{-λ}\frac{λ^x}{x!}$$

$$E(X^2)=\sum_{x=2}^{n}λ^2e^{-λ}\frac{λ^{x-2}}{(x-2)!} + E(X)$$

$$E(X^2)=λ^2e^{-λ}\sum_{x=2}^{n}\frac{λ^{x-2}}{(x-2)!} +λ$$

+ **Reorganizando**. Haciendo el cambio $k=x-2$

$$E(X^2)=λ^2e^{-λ}\sum_{k=0}^{n}\frac{λ^{k}}{k!} +λ$$
$$E(X^2)=λ^2\sum_{k=0}^{n}e^{-λ}\frac{λ^{k}}{k!} +λ$$
$$E(X^2)=λ^2(1) +λ=λ^2 +λ$$


+ **Varianza** $$V(X)=E(X-μ)^2=σ^2=E(X^2)- {\mu}^2 =[λ^2 +λ] -[λ]^2 \\ = λ$$



## **d. Función de Distribución Acumulada (FDA)** La función de Distribución acumulada para una variable $$X∼P(λ)$$ viene dada por

Una variable aleatoria discreta $X$ sigue una distribución de Poisson si tiene funcion FMP dada por la formula:
$$ P(X=x)= f(x,λ) = e^{-λ}\frac{λ^x}{x!},  x≥0$$

$$F(x)=P[X\leq x]= \sum_{k=o}^{x}e^{-λ}\frac{λ^k}{k!}$$

## **e. Otras Propiedades**

+ **Función Generadora de Momentos**
 $$M_X(t)=E(e^{tX})=\sum_{x=o}^{∞}e^{tx}e^{-λ}\frac{λ^x}{x!}$$
 $$M_X(t)=E(e^{tX})=e^{-λ}\sum_{x=o}^{∞}\frac{(e^tλ)^x}{x!}$$
 $$M_X(t)=E(e^{tX})=e^{-λ}e^{e^tλ} \ \
 =e^{-λ+e^tλ}=e^{(e^t-1)λ}$$

+ **Función Generadora de Probabilidad** $$G_X(t)=E(t^X)$$

$$G_X(t)=E(t^{X})=\sum_{x=o}^{∞}t^xe^{-λ}\frac{λ^x}{x!}=e^{-λ}\sum_{x=o}^{∞}\frac{(tλ)^x}{x!}=e^{-λ}e^{tλ}\ \
 =e^{-λ+tλ}=e^{(t-1)λ}$$


+ **Función característica** $$\Phi_X(t)=E(e^{itX})=e^{(e^{it}-1)λ}$$

## **f. Propiedades de la función generadora de momentos**

+ **Teorema de Unicidad**.  Si existe $M_X(t)$ de una v.a $X$, determna únivocamente su distribución de probabilidad.
+ **Teorema de relación con los momentos**. Si existe $M_X(t)$ de una v.a $X$ para $t∈(-t_0,t_1)$, se tiene:
+ + Existe el momento $E(X^k)$
+ + $$M_X(t)=\sum_{j=o}^{+∞}\frac{t^j}{j!}E(X^j), t∈(-t_0,t_1)$$
+ + $$E(X^k) =\frac{d^jM_X(t)}{dt^j}|_{t=0}$$

+ **Para variables aleatorias discretas, si existen $$M_X(t)=E(e^{tX}) \ y \ G_X(t)=E(t^{X})$$**, se cumple $$M_X(t)=E(e^{tX})=E((e^t)^{X})=\ G_X(e^t)$$

# **160. Aproximación Binomial a Poisson**


## **a. Teorema. Si n es “grande” y  p es “pequeño” tenemos la siguiente aproximación:**

$$ P(X=x)= f(x,p) = \begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}\sim e^{-λ}\frac{λ^x}{x!}= f(x,λ),\  λ=np$$

+ Dado que
$$\begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}=\frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)} \ \
=\frac{n}{x}\frac{(n-1)!}{(x-1)![(n-1)-(x-1)]!}p^x(1-p)^{(n-x)} \ \
 \ \
\ \
=\frac{n}{x}\begin{pmatrix} n-1 \\ x-1 \end{pmatrix}
=\frac{n}{x}\frac{(n-1)}{(x-1)}\frac{(n-2)}{(x-2)}\cdots\frac{(n-(x-1))}{(1)}p^x(1-p)^{(n-x)}$$

+ **Factorizando $n$**
$$\begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}=\frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)} \ \
=\frac{n^x}{x}\frac{(1-1/n)}{(x-1)}\frac{(1-2/n)}{(x-2)}\cdots\frac{(1-(x-1)/n)}{(1)}p^x(1-p)^{(n-x)}$$

+ **Factorizando $p =\frac{\lambda}{ n}$**
$$\begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}=\frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)} \ \
=\frac{n^x}{x}\frac{(1-1/n)}{(x-1)}\frac{(1-2/n)}{(x-2)}\cdots\frac{(1-(x-1)/n)}{(1)}(\frac{\lambda}{n})^x(1-\frac{\lambda}{n})^{(n-x)}$$

+ **Reorganizando**

$$\begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}=\frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)} \ \
={\lambda ^x}\frac{n^x}{x!n^x}(1-1/n)(1-2/n)\cdots(1-(x-1)/n)(1-\frac{\lambda}{n})^{n}(1-\frac{\lambda}{n})^{-x}$$

+ **Simplificando**

$$\begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)}=\frac{n!}{x!(n-x)!}p^x(1-p)^{(n-x)} \ \
=\frac{{\lambda ^x}}{x!}(1-1/n)(1-2/n)\cdots(1-(x-1)/n)(1-\frac{\lambda}{n})^{n}(1-\frac{\lambda}{n})^{-x}$$

+ **Tomando limite**
$$\lim \limits_{n\to\infty} \begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)} \ \
=\lim \limits_{n\to\infty} [\frac{{\lambda ^x}}{x!}(1-1/n)(1-2/n)\cdots(1-(x-1)/n)(1-\frac{\lambda}{n})^{n}(1-\frac{\lambda}{n})^{-x}]$$

$$\lim \limits_{n\to\infty} \begin{pmatrix} n \\ x \end{pmatrix}p^x(1-p)^{(n-x)} \ \
=\frac{{\lambda ^x}}{x!}\lim \limits_{n\to\infty} [(1-\frac{\lambda}{n})^{n}]=\frac{{\lambda ^x}}{x!}e^{-\lambda}$$


+ **Recordando** \\
$\lim \limits_{n\to\infty}=1/n=0$ \\
$\lim \limits_{n\to\infty} [(1-\frac{\lambda}{n})^{n}]=e^{-\lambda}$




# **b. Ejemplo de Aproximación Binomial a Poisson**

```python
# **b. Aproximación Binomial a Poisson**
plt.figure(figsize=(10, 6))
n, p = 1000, 0.005 # parametros de forma de la distribución binomial
mu =  5.1 # parametro de forma
poisson = stats.poisson(mu) # Distribución
x = np.arange(stats.binom.ppf(0.00001, n, p),
              stats.binom.ppf(0.999999, n, p))
x_p = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = stats.binom.pmf(x, n, p) # Función de Masa de Probabilidad
fmp_p = poisson.pmf(x_p) # Función de Masa de Probabilidad
ax.set_ylim(0, 0.2)
plt.plot(x, fmp, '--')
plt.plot(x_p, fmp_p)
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.vlines(x_p, 0, fmp_p, colors='r', lw=5, alpha=0.5)
plt.title('Función de Masa de Probabilidad Binomial vs Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

```

# **161. Aproximación Binomial a Poisson**

+ **Ejercicio 21.** *En cierta fábrica los accidentes ocurren con muy poca frecuencia. Se sabe que la probabilidad de un accidente en cualquier día dado es de 0.005, y que los accidentes son independientes entre sí.  a) ¿Cuál es la probabilidad de que en un día de cualquier periodo determinado de 400 días ocurra un accidente? b) ¿Cuál es la probabilidad de que ocurran a lo sumo tres accidentes en tal periodo?*

## **a. Solución**
+ $p=0.005$ es la probabilidad de que ocurra un accidente en cualquier día.
+ $n=400$ número de dias a observar
+ La v.a.d $X=$ $#$ de accidentes que ocurren en un periodo de $n=400$ dias.
+ **Distribución de Probabilidad**  $$X∼B(n,p)≃Y∼P(λ=np)$$
+ $λ=np=400*0.005=2$
+ La v.a.d $Y=$ $#$ de accidentes que ocurren en un dia cualquiera tiene promedio $λ=2$


# **b. Aproximación Binomial a Poisson**

```python
# **b. Aproximación Binomial a Poisson**
plt.figure(figsize=(10, 6))
n, p = 400, 0.005 # parametros de forma de la distribución binomial
mu =  2 # parametro de forma
poisson = stats.poisson(mu) # Distribución
x = np.arange(stats.binom.ppf(0.01, n, p),
              stats.binom.ppf(0.99, n, p))
x_p = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = stats.binom.pmf(x, n, p) # Función de Masa de Probabilidad
fmp_p = poisson.pmf(x_p) # Función de Masa de Probabilidad
ax.set_ylim(0, 0.2)
plt.plot(x, fmp, '--')
plt.plot(x_p, fmp_p)
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.vlines(x_p, 0, fmp_p, colors='r', lw=5, alpha=0.5)
plt.title('Función de Masa de Probabilidad Binomial vs Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```
