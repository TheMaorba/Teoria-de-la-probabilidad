# **177. Modelos Probabilísticos Continuos - La distribución Exponencial**

# **a. Generalidades de la Distribución Exponencial**

+ **Wikipedia**. En Teoría de Probabilidad y Estadística, la distribución exponencial es una distribución continua que se utiliza para modelar tiempos de espera para la ocurrencia de un cierto evento.
+ Esta distribución al igual que la distribución geométrica tiene la propiedad de pérdida de memoria.
+ La distribución exponencial es un caso particular de la distribución gamma.


+ **Ejemplos para la distribución exponencial**:
 - $X=:$ *El tiempo transcurrido entre dos llamadas en un centro de atención de llamadas.*
 - $X=:$ *El tiempo que debe esperar una persona hasta que pasa un taxi libre por una calle concreta.*
 - $X=:$ *El tiempo de espera hasta que entra un nuevo cliente en una tienda.*
 - $X=:$ *es la distribución de la longitud de los intervalos de una variable continua que transcurren entre dos sucesos, que se distribuyen según la distribución de Poisson.*

 - $X=:$ *El tiempo transcurrido en un centro de llamadas hasta recibir la primera llamada del día se podría modelar como una exponencial.*
  - $X=:$ *El intervalo de tiempo entre terremotos (de una determinada magnitud) sigue una distribución exponencial.*
 - $X=:$ *Supongamos una máquina que produce hilo de alambre, la cantidad de metros de alambre hasta encontrar una falla en el alambre se podría modelar como una exponencial.
  - $X=:$ *En fiabilidad de sistemas, un dispositivo con tasa de fallo constante sigue una distribución exponencial.*


# **b. Distribuciones continuas - La Distribución Exponencial**

+ Una variable aleatoria continua $X$ sigue una distribución *Exponencial* de parámetros $β$, lo cual denotamos por $$X∼exp(β)$$ si tiene funcion FDP dada por la formula:
$$ P(X=x)= f(x,β) = \frac{1}{β}exp{[\frac{-x}{β}]}, \ x\geq 0$$
+ Tambien $$E(X)=\mu=β,  \ \ \ \ V(X)=σ^2=β^2$$

+ Si $ \lambda$ representa al parámetro de una distribución de Poisson, entoces  $ \lambda= \frac{1}{β}$.
+ $\mu$ representa al número esperado.
+ $V(X)$ representa a la varianza.
+ Se puede mostrar


# **c. Valor Esperado de una Variable aleatoria normal**

+  $$E(X)=\mu$$
+ $$E(X) = ∫_{0}^{∞}xf(x)dx= ∫_{0}^{∞}\frac{x}{β}exp{[\frac{-x}{β}]}dx$$
+ **integrando por partes**
 - Hagamos $u=x →du=dx$ y $dv=\frac{1}{β}exp{[\frac{-x}{β}]}dx→v=-exp{[\frac{-x}{β}]}$
 - Aplicando la formula de integral por partes $$E(X) = ∫_{0}^{∞}xf(x)dx=-xexp{\{\frac{-x}{β}\}}|_{0}^{∞}+ ∫_{0}^{∞}exp{\{\frac{-x}{β}\}}dx$$
 $$E(X) =  ∫_{0}^{∞}exp{\{\frac{-x}{β}\}}dx=-βexp{\{\frac{-x}{β}\}}|_{0}^{∞}=β=\mu$$



# **d. Varianza de una Variable aleatoria normal**

+  $$V(X)=σ^2$$
 -  $$E(X^2) = ∫_{0}^{∞}x^2f(x)dx=  ∫_{0}^{∞}\frac{x^2}{β}exp{[\frac{-x}{β}]}dx$$


+ **integrando por partes**
 - Hagamos $u=x^2 →du=2xdx$ y $dv=\frac{1}{β}exp{[\frac{-x}{β}]}dx→v=-exp{\{\frac{-x}{β}\}}$
 - Aplicando la formula de integral por partes $$E(X^2) = ∫_{0}^{∞}x^2f(x)dx=-x^2exp{\{\frac{-x}{β}\}}|_{0}^{∞}+ 2∫_{0}^{∞}xexp{\{\frac{-x}{β}\}}dx$$
 $$E(X^2) = 2β ∫_{0}^{∞}\frac{x}{β}exp{\{\frac{-x}{β}\}}dx=2β(β)=2β^2$$

+ **Aplicando la formula de varianza**
 +  $$V(X)=σ^2 = E(X^2)-E(X)^2=2β^2-β^2=β^2$$

#**a. Construya el gráfico de la Distribución dada si  $μ=15$**

```python
# **a. Grafico manual de la función
from numpy.ma.core import log
# nuestra libreria
import numpy as np
import matplotlib.pyplot as plt
# Definimos una funcion
media = 15 # parametro

def lognormal1(x):
  return np.exp(-x/media)/media
# los valores que va a tomar x
x = np.linspace(0,100,1000) # espacio donde vamos a graficar la funcion
fig, ax = plt.subplots() #lugar donde vamos a estar graficando, ax objeto y subplot porque vamos hacer varios graficos
ax.plot(x,lognormal1(x)) # variable independiente y la funcion
plt.title('Distribución Exponencial β=15')
plt.ylabel('Probabilidad')
plt.xlabel('Dias de antelacion')
ax.grid() # le da mas elegancia al grafico
```

```python
#**a. Construya el gráfico de la Distribución dada si  $μ=15$**
plt.figure(figsize=(8, 8))
scale = 15 # parametro
exponencial = stats.expon(scale=15)
x = np.linspace(exponencial.ppf(0.01),
                exponencial.ppf(0.99), 1000)
fp = exponencial.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Exponencial β=15')
plt.ylabel('Probabilidad')
plt.xlabel('Dias de antelacion')
ax.grid() # le da mas elegancia al grafico
plt.show()
```

# **i. Calcule la probabilidad de que un viajero compre un billete con entre 10 y 20 dias  de antelación de la compañia B**

```python
#**i. Calcule la probabilidad de que un viajero compre un billete con entre 15 y 20 dias  de antelación de la compañia B**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 100, 401)

y = stats.expon(scale=15).pdf(x)
a = 10 # Valor de la variable a calcular a la izquierda
b = 20 # Valor de la variable a calcular a la izquierda
c = stats.expon(scale=15).cdf(b)-stats.expon(scale=15).cdf(a) # Probabilidad pedida
x1 = np.linspace(a, b, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.expon(scale=15).pdf(x1) , color="r")



ax.set_xlabel('Dias de antelacion')
ax.set_ylabel('Densidad')
plt.title('Distribución Exponencial β=15')
ax.text(a, 0.00001, a)
ax.text(b, 0.00001, b)
ax.text(15, 0.02, f'Area solicitada = {c:.4f}')
print('la probabilidad pedida es')
print('P{10=<X<=20} =',f'Area solicitada = {c:.4f}')

plt.show()
```

#**k. ¿Qué porcentaje compran tiquetes superiores a 25 dias de antelacion y menores a 5 dias en la compañia B?**

```python
#**k. ¿Qué porcentaje compran tiquetes superiores a 25 dias de antelacion y menores a 5 dias en la compañia B?**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 100, 401)

y = stats.expon(scale=15).pdf(x)

a = 5 # Valor de la variable a calcular a la izquierda
b = 25 # Valor de la variable a calcular a la izquierda
c = stats.expon(scale=20).cdf(a) # Probabilidad pedida
d = 1-stats.expon(scale=20).cdf(b)# Probabilidad pedida
e = c+d # Probabilidad pedida
x1 = np.linspace(0, a, 401)
x2 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.expon(scale=15).pdf(x1) , color="r")
ax.fill_between(x2, 0,stats.expon(scale=15).pdf(x2) , color="r")

ax.set_xlabel('Dias de antelacion')
ax.set_ylabel('Densidad')
plt.title('Distribución Exponencial β=20')
ax.text(a, 0.00001, a)
ax.text(b, 0.00001, b)

ax.text(a-5, 0.02, f'Prob1 = {c:.4f}')
ax.text(b+15, 0.005, f'Prob2 = {d:.4f}')

ax.text(40, 0.06, f'Area solicitada = {e:.4f}')
print('la probabilidad pedida es')
print('P{X>=25} +P{X<=5}=',f'Area solicitada = {e:.4f}')

plt.show()
```

# **l. Dada una distribución exponencial con $μ=10$, calcule el valor de k tal que $P(X>k) = 0.301$ en la compañia A**


```python
# **l. Dada una distribución exponencial con $μ=10$, calcule el valor de k tal que $P(X>k) = 0.301$ en la compañia A**
x = np.linspace(0, 100, 401)
y = stats.expon(scale=10).pdf(x)
a = round(stats.expon(scale=10).ppf(0.699),3)
fig, ax = plt.subplots(figsize=[6, 4])
c = 1 - stats.expon(scale=10).cdf(a)
x1 = np.linspace(a, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.expon(scale=10).pdf(x1) , color="r")

ax.text(a+20, 0.04, f'Área pedida ={c:.4f}', horizontalalignment='center')

ax.text(a, 0.00001, f'k = {a}')
plt.title('Distribución Exponencial β=10')
plt.ylabel('Probabilidad')
plt.xlabel('Dias de antelacion')
ax.grid() # le da mas elegancia al grafico
print('El tiempo mínimo de compra del paquete es a',a,'dias')
print(f'El área a la derecha de P(X>={a})=',f'{c:.4f}')
plt.show()
```

# **m. Dada una distribución exponencial  con $μ=15$, calcule el valor de k tal que P (k < X < 20) = 0.42**

```python
# **m. Dada una distribución exponencial  con $μ=15$, calcule el valor de k tal que P (k < X < 20) = 0.42**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(0, 100, 401)

y = stats.expon(scale=15).pdf(x)
a = 20 # Valor de la variable a calcular a la izquierda
b = stats.expon(scale=15).cdf(a)
c = stats.expon(scale=15).cdf(a) - 0.42
d = round(stats.expon(scale=15).ppf(c),2) # valor critico pedido
e = stats.expon(scale=15).cdf(d) # Probabilidad acumulada
x1 = np.linspace(0, d, 401)
x2 = np.linspace(d, a, 401)
x3 = np.linspace(a, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.expon(scale=15).pdf(x1) , color="r", alpha = 0.3)
ax.fill_between(x2, 0,stats.expon(scale=15).pdf(x2) , color="b", alpha =0.3)
ax.fill_between(x3, 0,stats.expon(scale=15).pdf(x3) , color="y", alpha =0.3)

ax.text(d, 0.00001, d)
ax.text(a, 0.00001, a)

ax.text(40, 0.005, f'Area1= {1-b:.4f}')
ax.text(d+5, 0.01, f'Area2= {c:.4f}')
ax.text(0.1, 0.005, f'Area3= {e:.4f}')
plt.title('Distribución Exponencial β=10')
plt.ylabel('Probabilidad')
plt.xlabel('Dias de antelacion')
ax.grid() # le da mas elegancia al grafico
print(f'Area solicitada = {c:.4f}')

print(f'El valor critico solicitado es k ={d}')
print(f'P(X<={d})={e:.4f}')
print(f'P(k=<X<=20)={b-e:.4f}')
print(f'P(X<=20)={b:.4f}')

plt.show()
```
