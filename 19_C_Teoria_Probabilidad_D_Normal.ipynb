# **b. Distribuciones continuas - La Distribución Normal**

+ Una variable aleatoria continua $X$ sigue una distribución Normal de parámetros $μ$ y $σ$, lo cual denotamos por $$X∼N(μ , σ)$$ si tiene funcion FDP dada por la formula:
$$ P(X=x)= f(x,μ , σ) = \frac{1}{\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](x-μ)^2},
\ x>0$$

+ **Veamos que es una función de densidad de Probabilidad**
+ Sea  $I = ∫_{-∞}^{∞}f(x)dx= ∫_{-∞}^{∞}\frac{1}{\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](x-μ)^2}dx$

+ El cambio de variable $ω = \frac{x-μ}{σ}$

+ Sea  $I = ∫_{-∞}^{∞}\frac{1}{\sqrt{2\piσ^2}}exp{[\frac{-ω^2}{2}]}σdω$

+ $I = ∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω$
+ **Entonces**
 - $I^2 = [∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-x^2}{2}]}dx][∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-y^2}{2}]}dy]$
 - $I^2 = [∫_{-∞}^{∞}\frac{1}{{2\pi}}exp{[\frac{-(x^2+y^2)}{2}]}dxdy]$

+ **Cambiando a coordenadas polares**
 - $x = rcos(θ), y = rsin(θ), x^2+y^2=r^2$

 - $I^2 = [∫_{0}^{2\pi}∫_{0}^{∞}\frac{1}{{2\pi}}exp{[\frac{-r^2}{2}]}rdrdθ]$

 - $I^2 = [{2\pi}∫_{0}^{∞}\frac{1}{{2\pi}}exp{[\frac{-r^2}{2}]}rdr]$


 - $I^2 = [∫_{0}^{∞}rexp\{\frac{-r^2}{2}\}dr]= - exp\{\frac{-r^2}{2}\}|_{0}^{∞}=1$
 + **Por tanto**
 $$I = ∫_{-∞}^{∞}f(x)dx= ∫_{-∞}^{∞}\frac{1}{\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](x-μ)^2}dx=1$$



# **c. Valor Esperado de una Variable aleatoria normal**

+  $$E(X)=\mu$$
+ $$E(X) = ∫_{-∞}^{∞}xf(x)dx= ∫_{-∞}^{∞}\frac{x}{\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](x-μ)^2}dx$$

+ El cambio de variable $ω = \frac{x-μ}{σ}$


+ $$E(X) = ∫_{-∞}^{∞}\frac{(σω+μ)}{\sqrt{2\piσ^2}}exp{[\frac{-ω^2}{2}]}σdω$$

+ **Separando las integrales**

+ $$E(X) = σ∫_{-∞}^{∞}\frac{ω}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω+ μ ∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω = σ(0)+ μ(1)$$

# **d. Varianza de una Variable aleatoria normal**

+  $$V(X)=σ^2$$
+ $$V(X)=σ^2=E(X-μ)^2 = ∫_{-∞}^{∞}(x-μ)^2f(x)dx= ∫_{-∞}^{∞}\frac{(x-μ)^2}{\sqrt{2\piσ^2}}exp{[\frac{-1}{2σ^2}](x-μ)^2}dx$$

+ El cambio de variable $ω = \frac{x-μ}{σ}$


+ $$V(X)=σ^2=E(X-μ)^2  = ∫_{-∞}^{∞}\frac{(σω)^2}{\sqrt{2\piσ^2}}exp{[\frac{-ω^2}{2}]}σdω$$

+ **Reorganizando la integral**

+ $$V(X)= σ^2∫_{-∞}^{∞}\frac{ω^2}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω$$


+ **Integrando por partes**

+ $$V(X)= σ^2∫_{-∞}^{∞}ω\frac{ω}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω$$

+ **haciendo** $u=ω$ y $dv=\frac{ω}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω$

+ **Se tiene**


+ $$V(X)= σ^2\{-ω\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}|_{-∞}^{∞}+∫_{-∞}^{∞}\frac{1}{\sqrt{2\pi}}exp{[\frac{-ω^2}{2}]}dω\} =σ^2\{0+1\}=σ^2 $$

```python
#**a. Construya el gráfico de la Distribución dada si los parámetros son μ=85,σ=3.8**
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy import stats
plt.figure(figsize=(8, 8))
#x = np.linspace(0, 20, 200)
media = 85 # parametro
sigma = 3.8 # parametro
normal = stats.norm(media,sigma )
x = np.linspace(normal.ppf(0.01),
                normal.ppf(0.99), 100)
fp = normal.pdf(x) # Función de Probabilidad
plt.plot(x, fp)
plt.title('Distribución Normal')
plt.ylabel('Probabilidad')
plt.xlabel('Duración')
plt.show()
```

## **Animacion de la distribución normal**

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy import stats
from IPython.display import HTML

# Parámetros de la distribución
mu, sd = 85, 3.8
x = np.linspace(70, 100, 401)
y = stats.norm(mu, sd).pdf(x)

# Valores objetivo
a = 70
b = 100
c = stats.norm(mu, sd).cdf(b) - stats.norm(mu, sd).cdf(a) # Probabilidad pedida

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=[6, 4])
ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
ax.set_title(f'P{{70 <= X <= 100}} = {c:.4f}, D. Normal con $μ=85,σ=3.8$')

# Inicialización de las líneas y áreas
line, = ax.plot(x, y, c='g')
fill = ax.fill_between([], [], color="r")
text_cdf = ax.text(87.25, 0.04, '', horizontalalignment='center')

# Función de inicialización para la animación
def init():
    line.set_data(x, y)
    return line,

# Número de cuadros necesarios para alcanzar el valor objetivo
num_frames = 100
incremento = (b - a) / num_frames

# Función de animación
def animate(i):
    global fill
    fill.remove()
    x1 = np.linspace(a, a + i * incremento, 401)
    fill = ax.fill_between(x1, 0, stats.norm(mu, sd).pdf(x1), color="r")
    probabilidad_actual = stats.norm(mu, sd).cdf(a + i * incremento) - stats.norm(mu, sd).cdf(a-0.25)
    text_cdf.set_text(f'{probabilidad_actual:.4f}')
    return line, fill, text_cdf

# Crear la animación
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=num_frames, interval=100, blit=True)

# Mostrar la animación
HTML(ani.to_html5_video())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy import stats
from IPython.display import HTML

# Datos iniciales
x = np.linspace(70, 100, 401)
y = stats.norm(loc=85, scale=3.8).pdf(x)
a = round(stats.norm(85, 3.8).ppf(0.785), 1)
target_value = stats.norm(loc=85, scale=3.8).ppf(1 - 0.215)

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=[6, 4])
ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
ax.set_title('Histograma Distribución Normal con $μ=85,σ=3.8$')

# Inicialización de las líneas y áreas
line, = ax.plot(x, y, c='g')
fill = ax.fill_between([], [], color="r")
text_cdf = ax.text(95, 0.05, '', horizontalalignment='center')
text_ppf = ax.text(88, 0.0001, '', horizontalalignment='center')

# Función de inicialización para la animación
def init():
    line.set_data(x, y)
    return line,

# Número de cuadros necesarios para alcanzar el valor objetivo
num_frames = int((100 - target_value) / 0.1)

# Función de animación
def animate(i):
    global fill
    current_x = 87.9 + i * (target_value - 87.9) / num_frames
    fill.remove()
    fill = ax.fill_between(x[x > current_x], y[x > current_x], color="r")
    text_cdf.set_text(f'{1 - stats.norm(loc=85, scale=3.8).cdf(current_x):.4f}')
    text_ppf.set_text(f'{a}')
    return line, fill, text_cdf, text_ppf

# Crear la animación
ani = animation.FuncAnimation(fig, animate, init_func=init, frames=num_frames, interval=100, blit=True)

# Mostrar la animación
HTML(ani.to_html5_video())
```

+ $μ=85,σ=3.8$
+ $$P\{|X-μ|\geq 2σ\}=P\{X-μ\geq 2σ\}+P\{-2σ\geq(X-μ)\} \ \
=P\{ X\geq μ+2σ\}+P\{X\leq μ-2σ\}=P\{ X\geq 92.6\}+P\{X\leq 77.4\}$$

```python
#**k. ¿Cuál es la probabilidad de que una batería dure más a lo más dos desviaciones estandard de su media?**
fig, ax = plt.subplots(figsize=[6, 4])
x = np.linspace(70, 100, 401)
mu, sd = 85, 3.8
y = stats.norm(mu, sd).pdf(x)
a = 77.4 # Valor de la variable a calcular a la izquierda
b = 92.6 # Valor de la variable a calcular a la izquierda
c = stats.norm(mu, sd).cdf(a) # Probabilidad acumulada
d = stats.norm(mu, sd).cdf(b) # Probabilidad acumulada
x1 = np.linspace(70, a, 401)
x2 = np.linspace(b, 100, 401)

ax.plot(x, y, c='g')

ax.fill_between(x1, 0,stats.norm(mu, sd).pdf(x1) , color="r")
ax.fill_between(x2, 0,stats.norm(mu, sd).pdf(x2) , color="b")

ax.text(b+0.1, 0.02, f'{1-d:.4f}', horizontalalignment='center')
ax.text(a-0.1, 0.02, f'{c:.4f}', horizontalalignment='center')

ax.set_xlabel('Valores de x')
ax.set_ylabel('Densidad')
plt.title('Histograma Distribución Normal con $μ=85,σ=3.8$')
print('la probabilidad de que una batería dure más a lo más dos desviaciones estandard de su media es')
print('P{88=<X8} =',c+1-d)

plt.show()
```
