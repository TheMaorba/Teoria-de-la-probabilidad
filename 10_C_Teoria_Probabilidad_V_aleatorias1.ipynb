# **Variables aleatorias y muestreo aleatorio - Conceptos**
## **a. Definición como función**

+ Recordemos que el espacio muestral $\Omega$, contiene todos los resultados posibles de un experimento aleatorio.
+ *Para cada uno de estos posibles resultados, $\omega$, una variable aleatoria asigna un valor (generalmente numérico), es decir, una variable aleatoria, $X$, es realmente una función $X: \Omega \rightarrow \mathbb{R}$, o $X: \Omega \rightarrow \mathbb{R}^n$*.

## **b. Definición formal**
+ La definición formal (es decir, desde el punto de vista de la teoría de la medida) de una variable aleatoria está dada por:

+ Dado un espacio de probabilidad $(\Omega, \mathcal{F}, \mathbb{P})$, una _variable aleatoria_ es una función $X$ de $\Omega$ a los números reales $\mathbb{R}$ tal que
$$
X(]\infty, x])^{-1} = \{\omega  \in \Omega: X(\omega) \leq x  \} \in \mathcal{F}, \ \mbox{para todo } x \in \mathbb{R}
$$
+ Se utilizará la siguiente notación
$$
\mathbb{P}_{X}[X(]\infty, x])^{-1}] :=\mathbb{P}[X \leq x]
$$
+ que se conoce como la definción inducida de probabilidad manera similar para los otros eventos.

## **Caso discreto**.
+  Una variable aleatoria $X$ con valores sobre el conjunto $\mathbb{R}$, es una función de $\Omega$ sobre un subconjunto finito o numerable infinito $\{x_1, x_2, \ldots\}$ de números reales, de tal manera que $X(]\infty, x])^{-1} =\{\omega: X(\omega) = x_i \} \in \mathcal{F}$ para toda $i$.

+ Cada variable aleatoria discreta, $X$, tiene asociada una función, $f$, llamada la _función de masa de probabilidad_.

+ Sea $X$ una variable aleatoria discreta con valores en $\mathbb{R}$. Si $f$ es una función tal que:
+ $f(x) = \mathbb{P}(X = x) \geq 0$ para todo $x \in \mathbb{R}$.
+ El conjunto $A = \{x: f(x) \neq 0 \}$ es un subconjunto finito o numerable infinito de $\mathbb{R}$.
+ $\sum_{x \in A} f(x) = 1$.
+ Entonces decimos que $f$ es una función de densidad (o función de masa) de $X$.

+  la variable aleatoria $X$ con distribución $f(x)=P\{X=x\}$

+ **Tambien se puede escribir**

$$f(x)= \left\{ \begin{array}{lcc}
             \dfrac{1}{4} &   si  & x=0, \\
             \\ \dfrac{1}{2} &   si  & x=1, \\
             \\ \dfrac{1}{4} &   si  & x=2,  \\
              \\ 0 &   en & otro & caso
             \end{array}
   \right.$$


A moneda2.png

```python
import itertools
import matplotlib.pyplot as plt

# Función para contar las caras en dos lanzamientos de una moneda
def contar_caras(numero_caras):
    resultados = itertools.product(['C', 'S'], repeat=numero_caras)
    distribucion = {}

    # Contar caras en cada combinación
    for resultado in resultados:
        caras = resultado.count('C')
        if caras in distribucion:
            distribucion[caras] += 1
        else:
            distribucion[caras] = 1

    # Calcular probabilidades
    n_lanzamientos = 2**2
    distribucion_probabilidades = {k: v / n_lanzamientos for k, v in distribucion.items()}

    return distribucion_probabilidades

# Calcular la distribución de probabilidad para 2 lanzamientos de una moneda
distribucion_probabilidades = contar_caras(2)

# Crear la figura y los ejes para el gráfico
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))



# Mostrar resultados
print("Número de caras | Probabilidad | x*f(x) |")
print("------------------------------------------")
for num_caras, probabilidad in distribucion_probabilidades.items():
    print(f"       {num_caras}        |    {probabilidad:.2f} |        {num_caras*probabilidad}")

# Gráfico de la distribución de probabilidad
ax1.bar(distribucion_probabilidades.keys(), distribucion_probabilidades.values())
ax1.set_title('Distribución de Probabilidad')
ax1.set_xlabel('Número de caras')
ax1.set_ylabel('Probabilidad')



# Histograma de frecuencias relativas
num_caras = list(distribucion_probabilidades.keys())
probabilidades = list(distribucion_probabilidades.values())
ax2.hist(num_caras, bins=[-0.5, 0.5, 1.5, 2.5, 3.5], weights=probabilidades, edgecolor='black', alpha=0.7)
ax2.set_title('Histograma de Frecuencias Relativas')
ax2.set_xlabel('Número de caras')
ax2.set_ylabel('Frecuencia relativa')

# Mostrar la figura
plt.tight_layout()
plt.show()
```

## **Mas sobre variables aleatorias Discretas**

+ Definamos $\Omega = \{x_1, x_2, \ldots \}$, $\mathcal{F} = \mathcal{P}(\Omega)$ (el conjunto potencia de $\Omega$) y $\mathbb{P}$ la medida de probabilidad definida sobre $\mathcal{F}$ como $\mathbb{P}( \{\omega \} ) = f(x_i)$ si $\omega = x_i$.
+  Si definimos $X$ como $X(\omega) = x_i$ si $\omega = x_i$, entonces
$$
\mathbb{P}(X = x_i) = \mathbb{P}(\{\omega: X(\omega) = x_i \}) = \mathbb{P}(\{x_i\}) = f(x_i)
$$
+ Finalmente sea $A \subseteq \mathbb{R}$ y sea $X$ una variable aleatoria discreta con posibles valores $x_1, x_2, \ldots$.
+ Entonces el conjunto $\{\omega: X(\omega) \in A  \}$ es un evento, esto se debe a que
$$
\{\omega: X(\omega) \in A  \} = \bigcup_{x_i \in A} \{\omega: X(\omega) = x_i \}
$$
+ Es común utilizar la siguiente notación

$$
\{\omega: X(\omega) \in A  \} = \{X \in A\}
$$
+ Podemos calcular $\mathbb{P}(X \in A)$ utilizando la siguiente fórmula
$$
\mathbb{P}(X \in A) = \sum_{x_i \in A}f(x_i)
$$
+ Demuestre que
$$
\mathbb{P}(X \in A) = \sum_{x_i \in A}f(x_i)
$$

## **Distribución acumulada - $F(t)$**
+ La función $F(t), -\infty < t < \infty$ definida por $$
F(t) = \mathbb{P}(X \leq t) = \sum_{x \leq t} f(x)$$
+ recibe el nombre de *función de distribución acumulativa* (o simplemente función de distribución) de la variable aleatoria discreta $X$.

## **Propiedades de la Distribución acumulada - $F(t)$**
+ Sea $X:\Omega \rightarrow \mathbb{R}$, una variable aleatoria. Una función $F(x)$ es la función de distribución de $X$ si y sólo si satisface las siguientes condiciones:
+ Para toda $x \in \mathbb{R}$,  $0 \leq F(x) \leq 1$.
+ $F(x) \rightarrow 0$ cuando $x \rightarrow -\infty$, y  $F(x) \rightarrow 1$ cuando $x \rightarrow \infty$.
+ Dado un número real $a$, $F(x) \downarrow F(a)$ cuando $x \downarrow a$ (continuidad por la derecha).
+ Para cualesquiera dos números reales $x < y$, $F(x) \leq F(y)$ (monótona creciente).
+ El teorema es válido para cualquier tipo de variables con valores en los reales (discretas, continuas o mixtas).




## **Ejemplo. Lanzamiento de una moneda tres veces - $X=$ Número de caras**

```python
# prompt: Costruyeme la tabla de la distribución acumulada para el lanzamiento de tres dados y me realizas su grafico

import itertools
import matplotlib.pyplot as plt

# Función para contar las caras en tres lanzamientos de una moneda
def contar_caras(numero_caras):
    resultados = itertools.product(['S', 'C'], repeat=numero_caras)
    distribucion = {}

    # Contar caras en cada combinación
    for resultado in resultados:
        caras = resultado.count('C')
        if caras in distribucion:
            distribucion[caras] += 1
        else:
            distribucion[caras] = 1

    # Calcular probabilidades
    n_lanzamientos = 2** numero_caras
    distribucion_probabilidades = {k: v / n_lanzamientos for k, v in distribucion.items()}

    return distribucion_probabilidades

# Calcular la distribución de probabilidad para 3 lanzamientos de una moneda
distribucion_probabilidades = contar_caras(3)

# Calcular la distribución acumulada
distribucion_acumulada = {}
suma_acumulada = 0
for num_caras, probabilidad in distribucion_probabilidades.items():
    suma_acumulada += probabilidad
    distribucion_acumulada[num_caras] = suma_acumulada

# Imprimir la tabla de distribución acumulada
print("Número de caras | Probabilidad | Probabilidad acumulada")
print("---------------------------------------")
for num_caras, probabilidad_acumulada in distribucion_acumulada.items():
    # Access the probability for the current number of heads
    probabilidad = distribucion_probabilidades[num_caras]
    print(f"       {num_caras}        |    {probabilidad:.4f}     |    {probabilidad_acumulada:.4f}")

# Gráfico de la distribución acumulada
plt.step(distribucion_acumulada.keys(), distribucion_acumulada.values(), where='post')
plt.title('Distribución Acumulada')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad acumulada')
plt.xticks(list(distribucion_acumulada.keys()))
plt.yticks([0, 0.25, 0.5, 0.75, 1])
plt.grid(True)
plt.show()
```

# **Esperanza y varianza de una variable aleatoria discreta**
### **a. Valor Esperado**
+ El concepto de esperanza de una variable aleatoria está relacionado con la idea de promediar los posibles valores que la variable puede tomar.
+ En lugar de utilizar un promedio común, en donde a cada posible valor se le da la misma ponderación, las ponderaciones son asignadas a través de la función de masa de probabilidad de la variable.
- Definamos $\Omega = \{x_1, x_2, \ldots \}$, $\mathcal{F} = \mathcal{P}(\Omega)$ (el conjunto potencia de $\Omega$) y $\mathbb{P}$ la medida de probabilidad definida sobre $\mathcal{F}$ como $\mathbb{P}( \{\omega \} ) = f(x_i)$ si $\omega = x_i$.



+ Sea $X$ una variable aleatoria discreta. Si $\sum_{i} |x_i|f(x_i) < \infty$, definimos la esperanza (media) de $X$ como
$$
\mu = \mathbb{E}(X) = \sum_{i}x_if(x_i)
$$
## **b. Propiedades del valor Esperado**
+ $\mathbb{E}(C)=C$.
+ Si $X,Y$ son variables aleatorias definidas sobre el mismo espacio $\Omega$, ambas con esperanza finita y si $\mathbb{P}(X \leq Y) = 1$, entonces $\mathbb{E}(X) \leq \mathbb{E}(Y)$.
+ Si $X$ tiene esperanza finita y si $\mathbb{P}(X \geq c) = 1$, entonces $E[X] \geq c$. De la misma forma, si $\mathbb{P}(X \leq c) = 1$, entonces $E[X] \leq c$.
+ $|E[X]| \leq E[|X|]$.
+ Si $Y = g(X)$, entonces $E[Y] = \sum_{i}g(x_i)f(x_i)$.
+ Si $X_1, X_2, \ldots, X_n$ son variables aleatorias definidas sobre el mismo espacio $\Omega$, con esperanza finita y si $c_1, \ldots, c_n$ son constantes, entonces
$$
E\left[ \sum_{i=1}^{n}c_{i}X_i \right] = \sum_{i=1}^{n}c_iE[X_i]
$$



### **b. Varianza**
$$ \sigma^2 =\mathbb{V}(X)=\sum_{x\in\Omega}(x-\mu)^2f(x)=\mathbb{E}(X^2)-\mu^2 $$

+ La desviación estándar de $X$ se define como $\sigma = \sqrt{\sigma^2}$.

### **Ejemplo: Hallar la media y varianza para la variable $X$ = Número de caras en el lanzamiento de tres monedas**

```python
# prompt: Hallar la media y varianza para la variable # de caras en el lanzamiento de tres monedas

# Calcular la esperanza (media)
esperanza = sum(num_caras * probabilidad for num_caras, probabilidad in distribucion_probabilidades.items())

# Calcular la varianza
varianza = sum((num_caras - esperanza)**2 * probabilidad for num_caras, probabilidad in distribucion_probabilidades.items())

print(f"La media o valor esperado de X, es E(X) = {esperanza}")
print(f"La Varianza  de X, es V(X) = {varianza}")

```


## **Ejemplo - Serie geométrica.** Suponga que la variable aleatoria $X$ tiene distribución de probabilidad

$$f(x)= \left\{ \begin{array}{lcc}
             (\dfrac{1}{2})^x &   si  & x=1,2,\cdots, \\
             \\ 0 &   en& otro&caso
             \end{array}
   \right.$$

### **$\bullet$** Hallar el $\mathbb{E}(X)$ y $\mathbb{V}(X)$

$$\mu =\mathbb{E}(X)=\sum_{x\in\Omega}xf(x)=\sum_{x=1}^{\infty}x (\dfrac{1}{2})^x=1*(\dfrac{1}{2})^1+2*(\dfrac{1}{2})^2+3*(\dfrac{1}{2})^3+\cdots+\\=\sum_{x=1}^{\infty}[\sum_{y=1}^{x}1] (\dfrac{1}{2})^x=\sum_{y=1}^{\infty}[\sum_{x=y}^{\infty}(\dfrac{1}{2})^x] \\ =\sum_{y=1}^{\infty}\dfrac{(\dfrac{1}{2})^y-(\dfrac{1}{2})^{\infty}}{1-\dfrac{1}{2}}
\\ =2\sum_{y=1}^{\infty}(\dfrac{1}{2})^y \\  =2\dfrac{(\dfrac{1}{2})^1-(\dfrac{1}{2})^{\infty}}{1-\dfrac{1}{2}} = 2
$$






$$\mathbb{E}(X^2)=\sum_{x\in\Omega}x^2f(x)=\sum_{x=1}^{\infty}x^2(\dfrac{1}{2})^x \\
=1^2*(\dfrac{1}{2})^1+2^2*(\dfrac{1}{2})^2+3^2*(\dfrac{1}{2})^3+\cdots+\\
=\sum_{x=1}^{\infty}[(x(x-1)+x)] (\dfrac{1}{2})^x \\
=\sum_{x=1}^{\infty}[x(x-1) (\dfrac{1}{2})^x]+\sum_{x=1}^{\infty}x (\dfrac{1}{2})^x \\
=2\sum_{x=1}^{\infty}[\sum_{y=1}^{x-1}y (\dfrac{1}{2})^x]+2 \\
=2*\dfrac{1}{2}\sum_{x=1}^{\infty}[\sum_{y=1}^{x-1}y (\dfrac{1}{2})^{x-1}]+2 \\
=\sum_{y=1}^{\infty}[\sum_{x=y}^{\infty}y (\dfrac{1}{2})^{x}]+2 \\
 =\sum_{y=1}^{\infty}y\dfrac{(\dfrac{1}{2})^y-(\dfrac{1}{2})^{\infty}}{1-\dfrac{1}{2}}+2
\\ =2\sum_{y=1}^{\infty}y(\dfrac{1}{2})^y +2\\  =2\mathbb{E}({X}) +2 = 2*2+2=6
$$

### **$\bullet$** Por tanto, $$\sigma^2=\mathbb{E}(X^2)-\mu^2= 6-4 =2$$



## **Ahora usando Python y ChatGPT**

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de probabilidad f(x) = 0.5^x
def probabilidad(x):
    return 0.5 ** x

# Calculando la distribución de probabilidad y la distribución acumulada
x_values = np.arange(1, 21)  # Calcularemos hasta x = 20
prob_values = [probabilidad(x) for x in x_values]
cdf_values = np.cumsum(prob_values)

# Media y varianza aproximadas
media = np.sum(x_values * prob_values)
varianza = np.sum((x_values - media)**2 * prob_values)

# Graficando la función de probabilidad y la distribución acumulada
plt.figure(figsize=(12, 6))

# Función de probabilidad
plt.subplot(121)
plt.bar(x_values, prob_values, label='Probabilidad')
plt.xlabel('x')
plt.ylabel('Probabilidad')
plt.title('Función de probabilidad')
plt.xticks(x_values)
plt.grid(True)

# Distribución acumulada
plt.subplot(122)
plt.step(x_values, cdf_values, where='mid', label='Distribución acumulada')
plt.xlabel('x')
plt.ylabel('Probabilidad acumulada')
plt.title('Distribución acumulada')
plt.xticks(x_values)
plt.grid(True)

plt.tight_layout()
plt.show()

# Mostrar la media y la varianza
print(f"La Media o valor Esperado de E es E(X) = {media:.2f}")
print(f"La Varianza de X es V(X) =  {varianza:.2f}")
```
